```python
# find the unique values from categor
for col in gold_data.select_dtypes(include='object').columns:
    print(col)
    print(gold_data[col].unique)
```

```python
#finding missing volues
features_na = [features for features in gold_data.columns if gold_data[features].isnull().sum()>0]
for feature in features_na:
            print(features, np.round(gold_data[feature].isnull().mean(),4),  '% missing values' )
else:
            print("no missing value found")
```

```python
#find features with one value
for column in gold_data.columns:
    print(column,gold_data[column].nunique())
```

```python
#categorical features
categorical_features=[feature for feature in gold_data.columns if ((gold_data[feature].dtypes=='O') & (feature not in ['GLD']))]
categorical_features
```

```python
for feature in categorical_features:
    print('The feature is {} and number of categories are {} '.format(feature,len(gold_data[feature].unique())))
```

```python
#Explore the numerical features
numerical_features = [feature for feature in gold_data.columns if ((gold_data[feature].dtypes != 'O') & (feature not in ['GLD']))]
print('Number of numerical variables :', len(numerical_features))
gold_data[numerical_features].head()
```

```python
discrete_feature=[feature for feature in numerical_features if len(gold_data[feature].unique())<25]
print("Discrete varibles count: {}".format(len(discrete_feature)))
```

```python
#find continous numericalfeatures
continuous_features=[feature for feature in numerical_features if feature not in discrete_feature+['GLD']]
print('continuous feature count{}'.format(len(continuous_features)))
```

```python
#Distribution of continous numerical features
plt.figure(figsize=(20,60), facecolor='white')
plotnumber =1
for continuous_feature in continuous_features:
    ax = plt.subplot(12,3,plotnumber)
    sns.distplot(gold_data[continuous_feature])
    plt.xlabel(continuous_feature)
    plotnumber+=1
plt.show()
```

```python
#relation between continous numerical features and labels
plt.figure(figsize=(20,60), facecolor='white')
plotnumber =1
for feature in continuous_features:
    data =gold_data.copy()
    ax = plt.subplot(12,3,plotnumber)
    plt.scatter(data[feature],data['GLD'])
    plt.xlabel(feature)
    plt.ylabel('GLD')
    plt.title(feature)
    plotnumber+=1
plt.show()
```

```python
#find outliers in numerical feature
plt.figure(figsize=(20,60), facecolor='white')
plotnumber =1
for numerical_feature in numerical_features:
    ax = plt.subplot(12,3,plotnumber)
    plt.boxplot(gold_data[numerical_feature])
    plt.xlabel(numerical_feature)
    plotnumber+=1
plt.show()
```

```python
correlation =gold_data.corr()
```

```python
plt.figure(figsize =(8,8))
sns.heatmap(correlation, cbar= True, square=True, fmt='.1f',annot=True, annot_kws={'size':8}, cmap='Blues' )
```

```python
print(correlation['GLD'])
```

```python
sns.distplot(gold_data['GLD'],color='green')
```

```python
x =gold_data.drop(['Date','GLD'],axis=1)
y =gold_data['GLD']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=2)
```

```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
regressor.fit(x_train,y_train)
```

```python
test_data_prediction =regressor.predict(x_test)
```

```python
print(test_data_prediction)
```

```python
error_score = metrics.r2_score(y_test, test_data_prediction)
print("R squared error  : " , error_score)
```

```python
y_test = list(y_test)
```

```python
plt.plot(y_test, color='blue', label='Actual Value')
plt.plot(test_data_prediction, color='green', label='predicted Vale')
plt.title('Actual price vs predicted price')
plt.xlabel('Number of  values')
plt.ylabel('GLD price')
plt.legend()
plt.show()
```
